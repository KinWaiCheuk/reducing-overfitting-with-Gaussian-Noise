```python
from sklearn.datasets import make_classification
import numpy as np
from keras.layers import Activation, Dense, GaussianNoise, Input
from keras.models import Model, Sequential
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer, MinMaxScaler
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X, y = make_classification(n_samples=1000, n_features=10, n_informative=5, n_redundant=5, n_repeated=0, 
                    n_classes=10, n_clusters_per_class=1, 
                    weights=None, flip_y=0, class_sep=1.0, hypercube=True, 
                    shift=0.0, scale=1.0, shuffle=False, random_state=None)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=101)
```

# Normalizing data

```python
normalizer = MinMaxScaler(feature_range=(-1,1))
X_train_norm = normalizer.fit_transform(X_train)
X_test_norm = normalizer.transform(X_test)
```

```python
lb = LabelBinarizer()
y_train_onehot = lb.fit_transform(y_train)
y_test_onehot = lb.transform(y_test)
```

# Creating a model that will overfit

# Using Norm

```python
model1 = Sequential()
model1.add(Dense(100,input_shape=(10,),activation='relu'))
model1.add(Dense(50, activation='relu'))
model1.add(Dense(20,activation='relu'))
model1.add(Dense(10,activation='softmax'))
model1.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])
history = model1.fit(X_train_norm,y_train_onehot, validation_data=(X_test_norm,y_test_onehot),epochs=200)
```

```python
##Plotting results
fig1 = plt.figure(figsize=(10,5))
ax1 = fig1.add_subplot(121)
ax1.set_xlabel('epoches')
ax1.set_ylabel('loss')
ax1.plot(history.history['loss'])
ax1.plot(history.history['val_loss'])
ax1.legend(['loss','val loss'])

ax2 = fig1.add_subplot(122,frameon=True)
ax2.set_xlabel('epoches')
ax2.set_ylabel('accuracy')
ax2.plot(history.history['acc'])
ax2.plot(history.history['val_acc'])
ax2.legend(['acc','val acc'])

fig1.suptitle('Fancy Model that Overfits', fontsize=16)
```

# Gaussian with Norm

```python
model2 = Sequential()
model2.add(GaussianNoise(0.08,input_shape=(10,)))
model2.add(Dense(100,activation='relu'))
model2.add(Dense(50, activation='relu'))
model2.add(Dense(20,activation='relu'))
model2.add(Dense(10,activation='softmax'))
model2.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])
history = model2.fit(X_train_norm,y_train_onehot, validation_data=(X_test_norm,y_test_onehot),epochs=200)
```

```python
##Plotting results
fig1 = plt.figure(figsize=(10,5))
ax1 = fig1.add_subplot(121)
ax1.set_xlabel('epoches')
ax1.set_ylabel('loss')
ax1.plot(history.history['loss'])
ax1.plot(history.history['val_loss'])
ax1.legend(['loss','val loss'])

ax2 = fig1.add_subplot(122,frameon=True)
ax2.set_xlabel('epoches')
ax2.set_ylabel('accuracy')
ax2.plot(history.history['acc'])
ax2.plot(history.history['val_acc'])
ax2.legend(['acc','val acc'])

fig1.suptitle('Using Gaussian Noise', fontsize=16)
```
